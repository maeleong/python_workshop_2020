## 4.4 Decision control statements

### 4.4.1 Conditional statements: If, If-else, If-elif-else

#### 4.4.1.1 If

The **"if"** statement is the simplest form of decision control statement, commonly used in decision-making.

It is like saying, _"if a is oil, then pour it into the pan"_, or _"if a is water, then pour it into the cylinder".

The syntax for writing an **if** statement begins with the letter **if**, followed by **your condition(s)**, and then a **colon (:)** symbol. Then, begin your _'operations'_ in the next line. Note that you must indent this line by at least 4 spaces.

Example:

if [condition]:

    statement 1
    
    statement 2
    

**Example 1:**

The **if** statement states that *if x is greater than 5*, then print *'x is greater than 0'*.

Since x is 6, it is greater than 5, so Python prints the statement *'x is greater than 0'*.

```python
x = 6

if x>5:
    print('x is greater than 0')
```

**Example 2:**

The first **if** statement states that *if x is equal to or less than 5*, then perform the computation of *10 - x* and print the value of *a*. Since x is 5, which satisfy the condition, then the computation will be performed and the value of *a* will be printed.

The second **if** statement states that *if x is greater than 6, then perform the computation of 10 + x*. However, since x is 5 and is less than 6, the second **if** statement is not performed.

```python
x = 5

if x<=5:
    a = 10 - x
    print(a)

if x>6:
    b = 10 + x
    print(b)
```

#### 4.4.1.2 if-else

The **if-else** statement is just as simple as the if statement. In addition to **if** statement, the **else** statement can be used to define an alternate block of statement to be executed if the expression in the **if** statement is **not true**.

The syntax for **if-else** statement is:

if [condition]:

    statement 1
    
else:

    statement 2

**Example**

In the example below, *y* is given the value of *-50*.

The conditional statements are *"if y is greater than zero, print this statement, otherwise, if y is less than zero, print the other statement."*

Since y is -50, it is less than zero. So, the **else** statement will be executed, and you get *y is less than 0' printed.

```python
y = -50
if y>0:
    print('y is greater than 0')
else:
    print('y is less than 0')
```

#### 4.4.1.3 if-elif-else

To create multiple conditions in the decision control statements, use **if-elif-else** statement. It works the same way as the if-else statement, except that it will go through several conditions in between **if-else** statements.

The syntax for **if-elif-else** statement is:

if [condition 1]:

    statement 1
    
elif [condition 2]:

    statement 2
    
elif [condition 3]:

    statement 3
    
else:

    statement 4

**Example:**

Here, *z* is given the value 50.

The **if-elif-else** statement states that
1. if *z* is equal to 0, then print *'z is zero'*.
2. if *z* is equal to 20, then print *'z is 20'*.
3. if *z* is equal to 50, then print *'z is 50'*.
4. if *z* does not satisfy any conditions above, then print *'z is not recognised'*.

Since z is 50, it satisfy the third condition, therefore, *'z is 50'* is printed.

```python
z = 50

if z==0:
    print('z is zero')
elif z==20:
    print('z is 20')
elif z==50:
    print('z is 50')
else:
    print('z is not recognised')
```

### 4.4.2 Iterative statements: For loop, while loop

Loop is a very popular programming function that repeats a given statement or a condition.

Suppose you have many oranges and you want to determine the pH of the oranges. You would repeat the same task of extracting juice sample from each orange in order to get the reading of the pH values.

If you're asking a robot to do that, you could write the instruction in computer code using the **for** loop.

### 4.4.2.1 For loop

To repeat a task until a given condition is met, we use the **for** loop.

The syntax for **for** loop is:

for *a variable* in *sequence*:

    statement 1

**Example:**
Say we have taken the pH values of 10 oranges and stored the pH values in a variable list called **ph_orange**, and we give a numerical ID for the oranges starting from 1 to 10 and stored these IDs in a variable list called **num_orange**.

Now, let's say, just for example's sake, we want to convert the pH values into percentage for all the oranges.

So, we write a **for** loop that iterates over the 10 oranges and converts the pH values into percentage values.

```python
ph_orange = [0.4,0.7,0.7,0.5,0.6,0.3,0.4,0.6,0.4,0.5]
num_orange = range(1,10)

for i in num_orange:
    pH_pct = ph_orange[i] * 100
    print(pH_pct)
```

#### 4.4.2.1 While loop

The **while** loop is similar to the **for** loop, but with a different syntax.

The syntax is somewhere similar to conditional statements:

while [condition]:
    
    statement 1
    
    statement 2
    
    statement 3

**Example:**

The example below gives a starting statement of num = 50.

The **while** loop states that *as long as the number **num** is less than 50, keep adding 5 to the number*.

**What it does is:**

With the starting number being 50, at 50, we add 5 and we get 55, then we repeat it again - 55 + 5 - and we get 60. The cycle repeats until the last number is 100 or less than 100. After 100, the loop stops because if we add 5 to 100, we get 105, which does not meet our given condition of **while num < 100**.


```python
num = 50

while num<100:
    num = num+5
    print(num)
```

## 4.6 Python functions

A function is a block of organised, reusable code that is used to perform a certain predefined task. Functions are an important part of any programming languages for two reasons:

1. They allow you to reuse the code youve written without needing to copy and modify the same code each time you want to use it.

2. They provide better modularity for your application, i.e. when you are doing a three-step process - such as search, compute and apply - for a main task for several different datasets, you can create these functions (search, compute, apply) and use and reuse these functions for the other datasets without having to copy and paste the block of codes into other datasets.

In Python, there are built-in functions such as print(), int(), float(), etc. which we have used in many of the examples in this session.

In the world of programming, we can also create our own functions. These functions are known as user-defined functions.

The syntax for defining a function is:

def function_name(parameters):

    statement1
    
    statement2
    
    return
    
It begins with the keyword **def**, which is short for define, followed by a name for the function and parameters or arguments (could be one or more) in the brackets. As in Decision Control statements, the next line should be **indented** and here is where we write our statements, *i.e.* perform a set of calculations. The function ends with **return** at the **same indentation** as the statements. This **return** keyword will return the output of the function.

**Example:**

In the example below, we create a function called **m_cm_conversion**, which is a function to convert metres into centimetres. And the parameter that is required as an input to this function is the height in metres.

In the next line, the function converts the height in metres into centimetres by multiplying it with 100. The new height in centimetres is stored in a variable called **height_cm**.

The last line of the function returns the value of the variable **height_cm** when you use the function.

```python
def m_cm_conversion(height_m):
    height_cm = height_m * 100
    return height_cm

height = m_cm_conversion(1.6)
print(height)
```

Let's try to apply a **for** loop into our function.

```python
heights_m = [1.5,1.6,1.7,1.9,2.0]

for i in heights_m:
    height_cm = m_cm_conversion(i)
    print(height_cm)
```

Below is an example using **for** loop and **indexing**.

```python
heights_m = [1.5,1.6,1.7,1.9,2.0]

for i in range(len(heights_m)):
    height_cm = m_cm_conversion(heights_m[i])
    print(height_cm)
```

#### Linear and multiple linear regression

#### Import statistics library

We will use a library called **statsmodels** and give it a shortname of **sm**.

In this linear regression, we will perform the Ordinary Least Square regression.

```python
import statsmodels.api as sm
```

#### Simple linear regression

```python
X = df[['RON95']]
X = sm.add_constant(X)
y = df['MYR']
model = sm.OLS(y,X,missing='drop') # this missing='drop' tells Python to exclude missing data in the analysis
result = model.fit()
print(result.summary())
```

#### Multiple linear regression

```python
X = df[['RON95','RON97','DIESEL']]
X = sm.add_constant(X)
y = df['MYR']
model = sm.OLS(y,X,missing='drop') # this missing='drop' tells Python to exclude missing data in the analysis
result = model.fit()
print(result.summary())
```
